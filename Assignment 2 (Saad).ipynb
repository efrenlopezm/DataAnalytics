```python
df_corpus_only_data = df_corpus.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1)
df_mugla_only_data = df_mugla.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1)
df_alex_only_data = df_alex.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1)
df_mag_only_data = df_mag.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1)
df_prag_only_data = df_prag.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1)
```

```python
df_list=[df_ankara, df_corpus, df_mugla, df_alex, df_mag, df_prag]
df = pd. concat(df_list, axis=1)
df
```

```python
new_df_list = []
for data in df_list:
  data['PARAMETER'] = data['PARAMETER'].replace(['PS'],'PRESSURE')
  data['PARAMETER'] = data['PARAMETER'].replace(['T2M'],'TEMPERATURE')
  data['PARAMETER'] = data['PARAMETER'].replace(['PRECTOTCORR'],'PRECIPITATION')
  data['PARAMETER'] = data['PARAMETER'].replace(['QV2M'],'HUMIDITY')
  data['PARAMETER'] = data['PARAMETER'].replace(['WS10M'],'WINDSPEED')
  data = data.drop('ANN', axis=1)
  new_df_list.append(data)
```

```python
final_data_list = []


for list in new_df_list:
  list.set_index('PARAMETER', inplace=True)

  pressure = list.loc['PRESSURE']
  temperature = list.loc['TEMPERATURE']
  humidity = list.loc['HUMIDITY']
  windspeed = list.loc['WINDSPEED']
  precipitation = list.loc['PRECIPITATION']

  variables = [pressure, temperature, humidity, 
               windspeed, precipitation]
  
  final_data_list.append(variables)
```

```python
data_dict = {'Ankara': final_data_list[0], 'Corpus': final_data_list[1], 'Mugla': final_data_list[2], 'Alex': final_data_list[3], 'Mag': final_data_list[4], 'Prag': final_data_list[5]}
```

```python
def flatten(l):
    return [item for sublist in l for item in sublist]
```

```python
var = ['pres', 'temp', 'hum', 'ws', 'prec']

# Ankara
ankara_data = {}
for i in range(5):
  rows = []
  for j in range(12):
    row = data_dict['Ankara'][i].iloc[j].tolist()
    row = [row[k] for k in range(1, len(row))]
    rows.append(row)
  ankara_data[f'ankara_{var[i]}'] = flatten(rows)
```

```python
# Corpus
corpus_data = {}
for i in range(5):
  rows = []
  for j in range(12):
    row = data_dict['Corpus'][i].iloc[j].tolist()
    row = [row[k] for k in range(1, len(row))]
    rows.append(row)
  corpus_data[f'corpus_{var[i]}'] = flatten(rows)

# mugla
mugla_data = {}
for i in range(5):
  rows = []
  for j in range(12):
    row = data_dict['Mugla'][i].iloc[j].tolist()
    row = [row[k] for k in range(1, len(row))]
    rows.append(row)
  mugla_data[f'mugla_{var[i]}'] = flatten(rows)

# Alex
alex_data = {}
for i in range(5):
  rows = []
  for j in range(12):
    row = data_dict['Alex'][i].iloc[j].tolist()
    row = [row[k] for k in range(1, len(row))]
    rows.append(row)
  alex_data[f'alex_{var[i]}'] = flatten(rows)


# Mag
mag_data = {}
for i in range(5):
  rows = []
  for j in range(12):
    row = data_dict['Mag'][i].iloc[j].tolist()
    row = [row[k] for k in range(1, len(row))]
    rows.append(row)
  mag_data[f'mag_{var[i]}'] = flatten(rows)


# Prag
prag_data = {}
for i in range(5):
  rows = []
  for j in range(12):
    row = data_dict['Prag'][i].iloc[j].tolist()
    row = [row[k] for k in range(1, len(row))]
    rows.append(row)
  prag_data[f'prag_{var[i]}'] = flatten(rows)
```

```python
#Hypothesis 7
#Wind velocity and humidity in coastal locations are inversely proportional

alex_hum = alex_data['alex_hum'] 
mag_hum = mag_data['mag_hum']
mugla_hum = mugla_data['mugla_hum'] 
corpus_hum = corpus_data['corpus_hum']



list_hum_cost = alex_hum + mag_hum + mugla_hum + corpus_hum
 
alex_ws = alex_data['alex_ws']
mag_ws = mag_data['mag_ws']
mugla_ws = mugla_data['mugla_ws']
corpus_ws = corpus_data['corpus_ws']

list_ws2_cost = alex_ws + mag_ws + mugla_ws + corpus_ws

len(list_ws2_cost)
```
